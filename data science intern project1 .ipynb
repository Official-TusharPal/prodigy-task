```python
import matplotlib.pyplot as plt

# Sample data - gender distribution in a population
genders = ['Male', 'Female', 'Non-binary']
counts = [4500, 5000, 300]

# Plotting the bar chart
plt.figure(figsize=(8, 6))
plt.bar(genders, counts, color=['blue', 'pink', 'purple'])

plt.xlabel('Gender')
plt.ylabel('Number of Individuals')
plt.title('Gender Distribution in Population')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random age data for illustration
np.random.seed(0)
ages = np.random.randint(18, 70, size=1000)

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=20, edgecolor='black', color='skyblue')

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution in Population')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data - age distribution in a population
ages = [18, 22, 25, 30, 35, 40, 45, 50, 55, 60]
counts = [1200, 1800, 2500, 3000, 2700, 2000, 1500, 1000, 600, 400]

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(ages, counts, width=3, color='skyblue')

plt.xlabel('Age')
plt.ylabel('Number of Individuals')
plt.title('Age Distribution in Population')
plt.xticks(ages)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
